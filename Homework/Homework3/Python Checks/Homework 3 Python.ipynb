```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Question 1 ###

#Initiate inputs, weights, and biases
p = np.linspace(-2,2,100)
w1 = [-1, 1]
b1 = [0.5, 1]
w2 = [1, 1]
b2 = -1

output = []

#Run through network
for i in range(0,len(p)):
    n11 = p[i]*w1[0] + b1[0]*1
    a11 = n11 if n11 > 0 else 0
    n12 = p[i]*w1[1] + b1[1]*1
    a12 = n12 if n12 > 0 else 0
    a21 = a11*w2[0] + a12*w2[1] + b2*1
    output.append(a21)
    
plt.plot(p, output)
plt.grid(color='lightgrey', linestyle='--', linewidth=1, alpha = 0.5)
plt.title('Homework 3, Question 1')
plt.xlabel('Input')
plt.ylabel('Network Response')
plt.ylim(bottom=0)
plt.show()
```

```python
### Question 2 ###

def hardlims(p, w, b):
    if(w*p+b < 0):
        a = -1
    else:
        a = 1
    return(a)

def hardlim(p, w, b):
    if(w*p+b < 0):
        a = 0
    else:
        a = 1
    return(a)

def purelin(p, w, b):
    a = w*p+b
    return(a)
        
def satlins(p, w, b):
    if(w*p+b < -1):
        a = -1
    elif(w*p+b > 1):
        a = 1
    else:
        a = w*p+b
    return(a)

def poslin(p, w, b):
    if(w*p+b < 0):
        a = 0
    else:
        a = w*p+b
    return(a)
    
    
def plot(w, b, pmin, pmax, act_func, question, subquestion, func):

    #Initiate inputs, weights, and biases
    p = np.linspace(pmin,pmax,100)
    output = []

    #Run through network
    for i in range(0,len(p)):
        a = act_func(p[i],w,b)
        output.append(a)

    plt.plot(p, output)
    plt.grid(color='lightgrey', linestyle='--', linewidth=1, alpha = 0.5)
    plt.title('Homework 3, Question {}.{}. {}'.format(question, subquestion, func))
    plt.xlabel('Input')
    plt.ylabel('Network Response')
    plt.ylim(top = max(output)+1, bottom=min(output)-1)
    plt.show()


plot(1, 1, -2, 2, hardlims, '2', 'i', 'hardlims')
plot(1, 1, -2, 2, hardlim, '2', 'ii', 'hardlim')
plot(2, 3, -2, 2, purelin, '2', 'iii', 'purelin')
plot(2, 3, -2, 2, satlins, '2', 'iv', 'satlins')
plot(2, 1, -2, 2, poslin, '2', 'v', 'poslin')

    
```

```python
### Question 3 ###

w111 = 2
w121 = 1
b11 = 2
b12 = -1
w211 = 1
w212 = -1
b21 = 0

output_n11 = []
output_n12 = []
output_a11 = []
output_a12 = []
output_n21 = []
output_a21 = []

p = np.linspace(-3,3,100)

for i in range(0, len(p)):
    n11 = p[i]*w111+b11
    n12 = p[i]*w121+b12  
    a11 = satlins(p[i],w111, b11)
    a12 = satlins(p[i],w121, b12)
    n21 = a11 * w211 + a12 * w212 + b21
    a21 = -1 if n21 < -1 else 1 if n21>1 else n21
    output_n11.append(n11)
    output_n12.append(n12)
    output_a11.append(a11)
    output_a12.append(a12)    
    output_n21.append(n21)
    output_a21.append(a21)
    
def net_plot(output, subquestion, var):
    plt.plot(p, output)
    plt.grid(color='lightgrey', linestyle='--', linewidth=1, alpha = 0.5)
    plt.title('Homework 3, Question 3.{}.'.format(subquestion))
    plt.xlabel('Input')
    plt.ylabel('Network Response for {}'. format(var))
    plt.ylim(top = max(output)+1, bottom=min(output)-1)
    plt.show()
    
net_plot(output_n11, 'i', 'n11')
net_plot(output_a11, 'ii', 'a11')
net_plot(output_n12, 'iii', 'n12')
net_plot(output_a12, 'iv', 'a12')
net_plot(output_n21, 'v', 'n21')
net_plot(output_a21, 'vi', 'a21')
```
